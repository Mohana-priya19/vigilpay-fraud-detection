```python
# VigilPay | Day 2 — Load & Explore Data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("✅ All libraries imported successfully!")
```

```python
# Load creditcard.csv from the data folder
df = pd.read_csv("../data/creditcard.csv")

print("✅ Dataset loaded!")
print(f"📐 Shape: {df.shape[0]} rows × {df.shape[1]} columns")
```

```python
df.head()
```

```python
missing = df.isnull().sum()
print("🚨 Missing Values:")
print(missing[missing > 0] if missing.sum() > 0 else "✅ No missing values found!")
```

```python
fraud_counts = df['Class'].value_counts()
fraud_percent = df['Class'].value_counts(normalize=True) * 100

print("🔢 Transaction Count:")
print(fraud_counts)
print(f"\n📊 Fraud %: {fraud_percent[1]:.4f}%")
print(f"📊 Legitimate %: {fraud_percent[0]:.4f}%")
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Class', data=df, palette=['#2ecc71', '#e74c3c'])
plt.title('VigilPay | Class Distribution\n(0 = Legitimate, 1 = Fraud)', fontsize=13)
plt.xticks([0, 1], ['Legitimate', 'Fraud'])
plt.xlabel('Transaction Type')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('../data/class_distribution.png')
plt.show()
print("✅ Chart saved!")
```
