```python
# VigilPay | Day 4 — Feature Engineering & Scaling
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Load dataset
df = pd.read_csv("../data/creditcard.csv")

print("✅ Libraries imported & dataset loaded!")
print(f"📐 Shape: {df.shape}")
```

```python
# Look at Amount and Time BEFORE scaling
print("📊 BEFORE SCALING:")
print(f"\n⏱️  Time  — Min: {df['Time'].min():.2f}  |  Max: {df['Time'].max():.2f}  |  Mean: {df['Time'].mean():.2f}")
print(f"💰 Amount — Min: {df['Amount'].min():.2f}  |  Max: {df['Amount'].max():.2f}  |  Mean: {df['Amount'].mean():.2f}")

print("""
🧠 WHY THIS IS A PROBLEM:
   • Time goes up to 172,792 — way too large!
   • Amount goes up to 25,691 — also too large!
   • V1-V28 are already scaled (done by the dataset creators)
   • We need to scale Time & Amount to match V1-V28
""")
```

```python
# Scale Amount and Time using StandardScaler
scaler = StandardScaler()

# Create new scaled columns (keep original ones too for reference)
df['scaled_Amount'] = scaler.fit_transform(df['Amount'].values.reshape(-1, 1))
df['scaled_Time']   = scaler.fit_transform(df['Time'].values.reshape(-1, 1))

# Check AFTER scaling
print("📊 AFTER SCALING:")
print(f"\n⏱️  scaled_Time   — Min: {df['scaled_Time'].min():.2f}  |  Max: {df['scaled_Time'].max():.2f}  |  Mean: {df['scaled_Time'].mean():.2f}")
print(f"💰 scaled_Amount  — Min: {df['scaled_Amount'].min():.2f}  |  Max: {df['scaled_Amount'].max():.2f}  |  Mean: {df['scaled_Amount'].mean():.2f}")

print("\n✅ Both features scaled successfully!")
```

```python
# Drop original Time and Amount (replaced by scaled versions)
df_clean = df.drop(columns=['Time', 'Amount'])

# Rename scaled columns for clarity
df_clean = df_clean.rename(columns={
    'scaled_Amount': 'Amount',
    'scaled_Time': 'Time'
})

# Reorder: put Class at the end
cols = [c for c in df_clean.columns if c != 'Class'] + ['Class']
df_clean = df_clean[cols]

print("✅ Clean dataset ready!")
print(f"📐 Shape: {df_clean.shape}")
print(f"\n📋 First 3 columns: {df_clean.columns[:3].tolist()}")
print(f"📋 Last 3 columns:  {df_clean.columns[-3:].tolist()}")
```

```python
# Separate features (X) and target (y)
X = df_clean.drop(columns=['Class'])  # everything except Class
y = df_clean['Class']                 # only Class column

print(f"✅ X shape (features): {X.shape}")
print(f"✅ y shape (target):   {y.shape}")

# Split: 80% training, 20% testing
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2,        # 20% for testing
    random_state=42,      # so results are reproducible
    stratify=y            # keeps same fraud ratio in both splits
)

print(f"\n📊 Training set:  {X_train.shape[0]} rows")
print(f"📊 Testing set:   {X_test.shape[0]} rows")

print(f"\n🔍 Fraud in training set: {y_train.sum()} ({y_train.mean()*100:.4f}%)")
print(f"🔍 Fraud in testing set:  {y_test.sum()} ({y_test.mean()*100:.4f}%)")
```

```python
# Save processed data to data folder
X_train.to_csv('../data/X_train.csv', index=False)
X_test.to_csv('../data/X_test.csv', index=False)
y_train.to_csv('../data/y_train.csv', index=False)
y_test.to_csv('../data/y_test.csv', index=False)

print("✅ All files saved to data/ folder!")
print("\n📁 Files created:")
print("   • X_train.csv — 227,845 rows × 30 features")
print("   • X_test.csv  — 56,962 rows  × 30 features")
print("   • y_train.csv — 227,845 labels")
print("   • y_test.csv  — 56,962 labels")
print("\n🎯 Ready for Day 5 — SMOTE & Model Training!")
```
